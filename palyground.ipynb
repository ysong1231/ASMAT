```python
import json
import requests
import matplotlib.pyplot as plt
```

```python
url = 'https://financialmodelingprep.com/api/v3/historical-chart/1min/%5EDJI'
s = requests.session()
request = s.get(url, timeout=15)
quote_data = request.json()
```

```python
quote_data
```

```python
x = []
y = []
for d in reversed(quote_data):
    if d['date'].split()[0] != '2020-04-14':
        continue
    x.append(d['date'])
    y.append(d['open'])
```

```python
plt.plot(x, y)
plt.show()
```
